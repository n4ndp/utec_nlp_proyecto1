```python
def load_and_tokenize(ds_key: str, tokenizer):
    cfg = DATASETS[ds_key]
    raw = load_dataset(*cfg["hf_args"])

    train_ds = raw["train"]
    val_ds   = raw[cfg["val_split"]]

    if cfg["max_train"] is not None:
        n = min(cfg["max_train"], len(train_ds))
        train_ds = train_ds.shuffle(seed=SEED).select(range(n))
    if cfg["max_val"] is not None:
        n = min(cfg["max_val"], len(val_ds))
        val_ds = val_ds.shuffle(seed=SEED).select(range(n))

    text_col, label_col = cfg["text_col"], cfg["label_col"]

    def _tok(batch):
        return tokenizer(batch[text_col], truncation=True, max_length=MAX_LEN)

    keep = {text_col, label_col}
    train_ds = train_ds.map(_tok, batched=True,
                            remove_columns=[c for c in train_ds.column_names if c not in keep])
    val_ds   = val_ds.map(_tok, batched=True,
                          remove_columns=[c for c in val_ds.column_names if c not in keep])

    if label_col != "labels":
        train_ds = train_ds.rename_column(label_col, "labels")
        val_ds   = val_ds.rename_column(label_col, "labels")

    train_ds = train_ds.remove_columns([text_col])
    val_ds   = val_ds.remove_columns([text_col])

    return train_ds, val_ds, cfg["num_labels"]


# sanity check con distilbert + sst2
_tok = AutoTokenizer.from_pretrained(MODELS["distilbert"])
_tr, _va, _nl = load_and_tokenize("sst2", _tok)
print("sst2 -> train:", len(_tr), " val:", len(_va), " num_labels:", _nl)
print("ejemplo:", {k: (v[:20] if hasattr(v, '__len__') else v) for k, v in _tr[0].items()})
```

```python
def compute_metrics(eval_pred):
    """métricas estándar para clasificación. weighted average para que
    funcione tanto en sst2/yelp (2 clases) como ag_news (4 clases)."""
    logits, labels = eval_pred
    preds = np.argmax(logits, axis=-1)
    acc = accuracy_score(labels, preds)
    prec, rec, f1, _ = precision_recall_fscore_support(
        labels, preds, average="weighted", zero_division=0
    )
    return {"accuracy": acc, "precision": prec, "recall": rec, "f1": f1}


def count_parameters(model):
    total = sum(p.numel() for p in model.parameters())
    trainable = sum(p.numel() for p in model.parameters() if p.requires_grad)
    return {"total_params": total, "trainable_params": trainable}


@torch.no_grad()
def measure_latency_and_memory(model, tokenizer, sample_texts, batch_size=1, n_warmup=10, n_runs=50):
    """mide latencia por sample y memoria gpu pico.
    batch_size=1 para latencia 'por inferencia' que es lo que pide el slide."""
    model.eval()
    device = next(model.parameters()).device

    # preparar un batch de muestra
    enc = tokenizer(sample_texts[:batch_size], padding=True, truncation=True,
                    max_length=MAX_LEN, return_tensors="pt").to(device)

    # warmup (la primera inferencia siempre es más lenta por cuda kernels)
    for _ in range(n_warmup):
        _ = model(**enc)

    # reset de stats de memoria
    if device.type == "cuda":
        torch.cuda.synchronize()
        torch.cuda.reset_peak_memory_stats()

    # latencia: usar cuda events para precisión
    if device.type == "cuda":
        starter = torch.cuda.Event(enable_timing=True)
        ender   = torch.cuda.Event(enable_timing=True)
        times_ms = []
        for _ in range(n_runs):
            starter.record()
            _ = model(**enc)
            ender.record()
            torch.cuda.synchronize()
            times_ms.append(starter.elapsed_time(ender))
        latency_ms = float(np.mean(times_ms))
        peak_mem_mb = torch.cuda.max_memory_allocated() / 1e6
    else:
        t0 = time.time()
        for _ in range(n_runs):
            _ = model(**enc)
        latency_ms = (time.time() - t0) * 1000 / n_runs
        peak_mem_mb = 0.0

    return {
        "latency_ms_per_inference": latency_ms,
        "peak_gpu_memory_mb": peak_mem_mb,
        "batch_size_used": batch_size,
    }
```

```python
class TransformerClassifier(nn.Module):
    """wrapper sobre distilbert/bert con un cabezal de clasificación configurable.
    diseñado para el ablation study:
      - freeze_transformer: congela todas las capas del transformer
      - hidden_layers: lista con neuronas por capa fc del cabezal.
        ej: [] -> solo linear final (más simple)
            [768] -> 1 capa oculta de 768
            [512, 256] -> 2 capas ocultas
    """
    def __init__(self, base_model_name, num_labels,
                 hidden_layers=None, dropout=0.1, freeze_transformer=False):
        super().__init__()
        self.base_model_name = base_model_name
        self.is_distilbert = "distilbert" in base_model_name.lower()
        self.config = AutoConfig.from_pretrained(base_model_name)
        self.transformer = AutoModel.from_pretrained(base_model_name)
        self.num_labels = num_labels
        self.hidden_layers = hidden_layers or []
        self.freeze_transformer = freeze_transformer

        if freeze_transformer:
            for p in self.transformer.parameters():
                p.requires_grad = False

        hidden_size = self.config.hidden_size  # 768 en ambos
        layers = []
        prev = hidden_size
        for h in self.hidden_layers:
            layers.append(nn.Linear(prev, h))
            layers.append(nn.GELU())
            layers.append(nn.Dropout(dropout))
            prev = h
        layers.append(nn.Linear(prev, num_labels))
        self.classifier = nn.Sequential(*layers)

    def forward(self, input_ids=None, attention_mask=None,
                token_type_ids=None, labels=None, **kwargs):
        # distilbert no acepta token_type_ids, hay que filtrarlo
        model_inputs = {"input_ids": input_ids, "attention_mask": attention_mask}
        if not self.is_distilbert and token_type_ids is not None:
            model_inputs["token_type_ids"] = token_type_ids

        outputs = self.transformer(**model_inputs)
        # tomamos la representación del [CLS] (primer token) para clasificar.
        # es consistente entre bert y distilbert (distilbert no tiene pooler).
        cls_repr = outputs.last_hidden_state[:, 0]
        logits = self.classifier(cls_repr)

        loss = None
        if labels is not None:
            loss = nn.CrossEntropyLoss()(logits, labels)

        return {"loss": loss, "logits": logits}


# sanity check: construir un modelo y ver el conteo de params
_m = TransformerClassifier(MODELS["distilbert"], num_labels=2, hidden_layers=[])
print("distilbert + head vacío:", count_parameters(_m))

_m2 = TransformerClassifier(MODELS["distilbert"], num_labels=2,
                            hidden_layers=[256, 128], freeze_transformer=True)
print("distilbert frozen + head [256,128]:", count_parameters(_m2))
del _m, _m2; gc.collect()
```

```python
class LossHistoryCallback(TrainerCallback):
    """captura train_loss y eval_loss en cada log step.
    es lo que pide el slide: 'Número de iteraciones vs (Training loss & Validation loss)'."""
    def __init__(self):
        self.train_history = []  # list of {step, loss}
        self.eval_history  = []  # list of {step, eval_loss, ...metrics}

    def on_log(self, args, state, control, logs=None, **kwargs):
        if logs is None:
            return
        step = state.global_step
        if "loss" in logs:
            self.train_history.append({"step": step, "loss": logs["loss"]})
        if "eval_loss" in logs:
            entry = {"step": step}
            for k, v in logs.items():
                if k.startswith("eval_"):
                    entry[k] = v
            self.eval_history.append(entry)
```

```python
@dataclass
class RunConfig:
    name: str
    base_model: str            # "distilbert" o "bert"
    dataset: str               # "sst2", "ag_news", "yelp"
    hidden_layers: List[int] = field(default_factory=list)
    freeze_transformer: bool = False
    epochs: int = EPOCHS
    batch_size: int = BATCH_SIZE
    lr: float = LR


def run_experiment(cfg: RunConfig, save_dir=None):
    """corre un experimento end-to-end y devuelve métricas + historia de loss."""
    print(f"\n{'='*70}\nrun: {cfg.name}\n{'='*70}")
    print(asdict(cfg))

    model_name = MODELS[cfg.base_model]
    tokenizer = AutoTokenizer.from_pretrained(model_name)

    train_ds, val_ds, num_labels = load_and_tokenize(cfg.dataset, tokenizer)
    print(f"datos -> train: {len(train_ds)}  val: {len(val_ds)}  labels: {num_labels}")

    model = TransformerClassifier(
        base_model_name=model_name,
        num_labels=num_labels,
        hidden_layers=cfg.hidden_layers,
        freeze_transformer=cfg.freeze_transformer,
    )
    params = count_parameters(model)
    print(f"params -> total: {params['total_params']:,}  trainable: {params['trainable_params']:,}")

    out_dir = f"/tmp/runs/{cfg.name}"
    os.makedirs(out_dir, exist_ok=True)

    args = TrainingArguments(
        output_dir=out_dir,
        num_train_epochs=cfg.epochs,
        per_device_train_batch_size=cfg.batch_size,
        per_device_eval_batch_size=EVAL_BATCH,
        learning_rate=cfg.lr,
        weight_decay=WEIGHT_DECAY,
        warmup_ratio=WARMUP_RATIO,
        bf16=True,                       # l4 soporta bf16 nativo
        eval_strategy="steps",
        eval_steps=200,
        logging_strategy="steps",
        logging_steps=50,
        save_strategy="no",              # no guardamos checkpoints para ahorrar disco
        report_to="none",
        seed=SEED,
        dataloader_num_workers=2,
    )

    loss_cb = LossHistoryCallback()
    collator = DataCollatorWithPadding(tokenizer)

    trainer = Trainer(
        model=model,
        args=args,
        train_dataset=train_ds,
        eval_dataset=val_ds,
        processing_class=tokenizer,
        data_collator=collator,
        compute_metrics=compute_metrics,
        callbacks=[loss_cb],
    )

    t0 = time.time()
    trainer.train()
    train_time_sec = time.time() - t0

    final_metrics = trainer.evaluate()
    print("eval final:", {k: round(v, 4) for k, v in final_metrics.items() if isinstance(v, (int, float))})

    # medir latencia y memoria con muestras reales del val set
    raw = load_dataset(*DATASETS[cfg.dataset]["hf_args"])
    text_col = DATASETS[cfg.dataset]["text_col"]
    val_split = DATASETS[cfg.dataset]["val_split"]
    sample_texts = [raw[val_split][i][text_col] for i in range(16)]
    eff = measure_latency_and_memory(model, tokenizer, sample_texts, batch_size=1)
    print("eficiencia:", {k: round(v, 3) for k, v in eff.items() if isinstance(v, (int, float))})

    result = {
        "config": asdict(cfg),
        "params": params,
        "metrics": {k: float(v) for k, v in final_metrics.items() if isinstance(v, (int, float))},
        "efficiency": eff,
        "train_time_sec": train_time_sec,
        "train_history": loss_cb.train_history,
        "eval_history": loss_cb.eval_history,
    }

    if save_dir:
        os.makedirs(save_dir, exist_ok=True)
        with open(os.path.join(save_dir, f"{cfg.name}.json"), "w") as f:
            json.dump(result, f, indent=2)
        print(f"guardado -> {save_dir}/{cfg.name}.json")

    # liberar gpu
    del model, trainer
    gc.collect()
    if torch.cuda.is_available():
        torch.cuda.empty_cache()

    return result


# ===== SANITY CHECK =====
# corrida corta: distilbert en sst2 con 1 epoch para verificar que todo funciona.
# debería tardar ~3-5 min en l4. si falla, no avanzamos a las fases reales.
_sanity_cfg = RunConfig(
    name="sanity_distilbert_sst2",
    base_model="distilbert",
    dataset="sst2",
    epochs=1,
)
_sanity_result = run_experiment(_sanity_cfg, save_dir=RESULTS_DIR)
print("\nsanity acc:", _sanity_result["metrics"].get("eval_accuracy"))
```

```python
# fase a: comparar distilbert vs bert con la config default (sin ablation)
# 6 corridas = 2 modelos × 3 datasets. ~50-60 min en l4.

PHASE_A_DIR = os.path.join(RESULTS_DIR, "phase_a_baseline")

phase_a_configs = []
for model_key in ["distilbert", "bert"]:
    for ds in ["sst2", "ag_news", "yelp"]:
        phase_a_configs.append(RunConfig(
            name=f"baseline_{model_key}_{ds}",
            base_model=model_key,
            dataset=ds,
            hidden_layers=[],          # solo linear final, default
            freeze_transformer=False,
            epochs=EPOCHS,
        ))

phase_a_results = {}
for cfg in phase_a_configs:
    res = run_experiment(cfg, save_dir=PHASE_A_DIR)
    phase_a_results[cfg.name] = res

# tabla resumen
rows = []
for name, r in phase_a_results.items():
    rows.append({
        "run": name,
        "model": r["config"]["base_model"],
        "dataset": r["config"]["dataset"],
        "accuracy": r["metrics"].get("eval_accuracy"),
        "f1": r["metrics"].get("eval_f1"),
        "precision": r["metrics"].get("eval_precision"),
        "recall": r["metrics"].get("eval_recall"),
        "params_M": r["params"]["total_params"] / 1e6,
        "latency_ms": r["efficiency"]["latency_ms_per_inference"],
        "peak_mem_MB": r["efficiency"]["peak_gpu_memory_mb"],
        "train_time_s": r["train_time_sec"],
    })
df_a = pd.DataFrame(rows)
print("\n=== FASE A: BASELINE ===")
print(df_a.to_string(index=False))
df_a.to_csv(os.path.join(PHASE_A_DIR, "summary.csv"), index=False)
```

```python
# fase b: ablation con distilbert en sst2. 4 configuraciones que cubren
# exactamente lo que pide el slide:
#   1. freeze del transformer (entrenar solo cabezal)
#   2. cabezal más ancho (más neuronas por capa)
#   3. cabezal más profundo (más capas)
#   4. cabezal más estrecho/chico (reducir neuronas)
# ~25-30 min en l4.

PHASE_B_DIR = os.path.join(RESULTS_DIR, "phase_b_ablation")

ablation_configs = [
    RunConfig(
        name="ablation_1_frozen",
        base_model="distilbert", dataset="sst2",
        hidden_layers=[256],          # cabezal mínimo necesario al congelar
        freeze_transformer=True,
    ),
    RunConfig(
        name="ablation_2_wider_head",
        base_model="distilbert", dataset="sst2",
        hidden_layers=[768],          # 1 capa oculta ancha
        freeze_transformer=False,
    ),
    RunConfig(
        name="ablation_3_deeper_head",
        base_model="distilbert", dataset="sst2",
        hidden_layers=[512, 256],     # 2 capas ocultas
        freeze_transformer=False,
    ),
    RunConfig(
        name="ablation_4_narrow_head",
        base_model="distilbert", dataset="sst2",
        hidden_layers=[128],          # capa pequeña
        freeze_transformer=False,
    ),
]

phase_b_results = {}
for cfg in ablation_configs:
    res = run_experiment(cfg, save_dir=PHASE_B_DIR)
    phase_b_results[cfg.name] = res

# tabla
rows = []
for name, r in phase_b_results.items():
    rows.append({
        "run": name,
        "head": r["config"]["hidden_layers"],
        "frozen": r["config"]["freeze_transformer"],
        "accuracy": r["metrics"].get("eval_accuracy"),
        "f1": r["metrics"].get("eval_f1"),
        "trainable_params_M": r["params"]["trainable_params"] / 1e6,
        "latency_ms": r["efficiency"]["latency_ms_per_inference"],
        "train_time_s": r["train_time_sec"],
    })
df_b = pd.DataFrame(rows)
print("\n=== FASE B: ABLATION ===")
print(df_b.to_string(index=False))
df_b.to_csv(os.path.join(PHASE_B_DIR, "summary.csv"), index=False)

# escoger mejor por f1
best_name = df_b.loc[df_b["f1"].idxmax(), "run"]
best_cfg_b = next(c for c in ablation_configs if c.name == best_name)
print(f"\nMEJOR CONFIG DEL ABLATION: {best_name}")
print(f"  hidden_layers: {best_cfg_b.hidden_layers}")
print(f"  freeze_transformer: {best_cfg_b.freeze_transformer}")
```

```python
# fase c: tomar la mejor config del ablation y compararla aplicada a
# distilbert y bert en los 3 datasets. esto es lo que pide el slide:
# "comparar únicamente la mejor versión ahora utilizando el modelo BERT como base.
#  Mostrar aquí todas las métricas."
# 6 corridas = 2 modelos × 3 datasets. ~50-60 min en l4.

PHASE_C_DIR = os.path.join(RESULTS_DIR, "phase_c_best_config")

phase_c_configs = []
for model_key in ["distilbert", "bert"]:
    for ds in ["sst2", "ag_news", "yelp"]:
        phase_c_configs.append(RunConfig(
            name=f"best_{model_key}_{ds}",
            base_model=model_key,
            dataset=ds,
            hidden_layers=best_cfg_b.hidden_layers,
            freeze_transformer=best_cfg_b.freeze_transformer,
            epochs=EPOCHS,
        ))

phase_c_results = {}
for cfg in phase_c_configs:
    res = run_experiment(cfg, save_dir=PHASE_C_DIR)
    phase_c_results[cfg.name] = res

# tabla
rows = []
for name, r in phase_c_results.items():
    rows.append({
        "run": name,
        "model": r["config"]["base_model"],
        "dataset": r["config"]["dataset"],
        "accuracy": r["metrics"].get("eval_accuracy"),
        "f1": r["metrics"].get("eval_f1"),
        "precision": r["metrics"].get("eval_precision"),
        "recall": r["metrics"].get("eval_recall"),
        "params_M": r["params"]["total_params"] / 1e6,
        "latency_ms": r["efficiency"]["latency_ms_per_inference"],
        "peak_mem_MB": r["efficiency"]["peak_gpu_memory_mb"],
        "train_time_s": r["train_time_sec"],
    })
df_c = pd.DataFrame(rows)
print("\n=== FASE C: MEJOR CONFIG ===")
print(df_c.to_string(index=False))
df_c.to_csv(os.path.join(PHASE_C_DIR, "summary.csv"), index=False)
```

```python
import matplotlib.pyplot as plt

PLOTS_DIR = os.path.join(RESULTS_DIR, "plots")
os.makedirs(PLOTS_DIR, exist_ok=True)

# ============================================================
# plot 1: bubble chart - params vs accuracy (lo pide explícito)
# ============================================================
# usamos fase a (baseline) que es la comparación más limpia
fig, ax = plt.subplots(figsize=(10, 6))

colors = {"distilbert": "#1f77b4", "bert": "#d62728"}
markers = {"sst2": "o", "ag_news": "s", "yelp": "^"}

for _, row in df_a.iterrows():
    # tamaño de burbuja = latencia (más grande = más lento)
    size = row["latency_ms"] * 60
    ax.scatter(row["params_M"], row["accuracy"],
               s=size, alpha=0.6,
               c=colors[row["model"]],
               marker=markers[row["dataset"]],
               edgecolors="black", linewidth=1.2)
    ax.annotate(f"{row['model'][:4]}-{row['dataset']}",
                (row["params_M"], row["accuracy"]),
                xytext=(8, 8), textcoords="offset points", fontsize=9)

ax.set_xlabel("Number of parameters (M)", fontsize=12)
ax.set_ylabel("Accuracy", fontsize=12)
ax.set_title("Parameters vs Accuracy (bubble size = inference latency)", fontsize=13)
ax.grid(True, alpha=0.3)

# leyenda manual
from matplotlib.lines import Line2D
legend_elems = [
    Line2D([0],[0], marker='o', color='w', markerfacecolor=colors["distilbert"],
           markersize=10, label='DistilBERT'),
    Line2D([0],[0], marker='o', color='w', markerfacecolor=colors["bert"],
           markersize=10, label='BERT'),
    Line2D([0],[0], marker='o', color='gray', markersize=10, label='SST-2', linestyle=''),
    Line2D([0],[0], marker='s', color='gray', markersize=10, label='AG News', linestyle=''),
    Line2D([0],[0], marker='^', color='gray', markersize=10, label='Yelp', linestyle=''),
]
ax.legend(handles=legend_elems, loc="lower right")

plt.tight_layout()
plt.savefig(os.path.join(PLOTS_DIR, "bubble_params_vs_accuracy.png"), dpi=200)
plt.savefig(os.path.join(PLOTS_DIR, "bubble_params_vs_accuracy.pdf"))
plt.show()


# ============================================================
# plot 2: training & validation loss vs steps
# ============================================================
# uno por dataset, mostrando distilbert vs bert (fase a)
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

for ax, ds in zip(axes, ["sst2", "ag_news", "yelp"]):
    for model_key in ["distilbert", "bert"]:
        name = f"baseline_{model_key}_{ds}"
        r = phase_a_results[name]

        train_steps = [x["step"] for x in r["train_history"]]
        train_loss  = [x["loss"] for x in r["train_history"]]
        eval_steps  = [x["step"] for x in r["eval_history"]]
        eval_loss   = [x["eval_loss"] for x in r["eval_history"]]

        c = colors[model_key]
        ax.plot(train_steps, train_loss, c=c, alpha=0.6,
                label=f"{model_key} train", linewidth=1.5)
        ax.plot(eval_steps, eval_loss, c=c, linestyle="--", marker="o",
                label=f"{model_key} val", linewidth=1.5, markersize=5)

    ax.set_title(f"{ds.upper()}")
    ax.set_xlabel("Step")
    ax.set_ylabel("Loss")
    ax.legend(fontsize=9)
    ax.grid(True, alpha=0.3)

plt.suptitle("Training & Validation Loss vs Steps (Phase A: Baseline)", fontsize=13)
plt.tight_layout()
plt.savefig(os.path.join(PLOTS_DIR, "loss_curves_baseline.png"), dpi=200)
plt.savefig(os.path.join(PLOTS_DIR, "loss_curves_baseline.pdf"))
plt.show()


# ============================================================
# plot 3: ablation comparison
# ============================================================
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# subplot 1: accuracy por config
ax = axes[0]
names = [c.name.replace("ablation_", "") for c in ablation_configs]
accs  = [phase_b_results[c.name]["metrics"]["eval_accuracy"] for c in ablation_configs]
bars = ax.bar(names, accs, color=["#888", "#1f77b4", "#2ca02c", "#ff7f0e"])
ax.set_ylim(0.80, 0.92)
ax.set_ylabel("Accuracy (SST-2 val)")
ax.set_title("Ablation: classifier head configurations")
ax.grid(True, alpha=0.3, axis="y")
for bar, acc in zip(bars, accs):
    ax.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 0.002,
            f"{acc:.3f}", ha="center", fontsize=10)

# subplot 2: loss curves de las 4 ablations
ax = axes[1]
ab_colors = ["#888", "#1f77b4", "#2ca02c", "#ff7f0e"]
for cfg, c in zip(ablation_configs, ab_colors):
    r = phase_b_results[cfg.name]
    eval_steps = [x["step"] for x in r["eval_history"]]
    eval_loss  = [x["eval_loss"] for x in r["eval_history"]]
    ax.plot(eval_steps, eval_loss, marker="o", label=cfg.name.replace("ablation_", ""),
            color=c, linewidth=1.8)
ax.set_xlabel("Step"); ax.set_ylabel("Validation loss")
ax.set_title("Ablation: validation loss curves")
ax.legend(fontsize=9); ax.grid(True, alpha=0.3)

plt.tight_layout()
plt.savefig(os.path.join(PLOTS_DIR, "ablation_comparison.png"), dpi=200)
plt.savefig(os.path.join(PLOTS_DIR, "ablation_comparison.pdf"))
plt.show()


print("\nplots guardados en:", PLOTS_DIR)
```
